# Module 3
# Data Formatting

Let us observe the types of each column in our dataset.

```python
df.dtypes
```

Not all the columns are of the specified/needed type. Let's change these.

```python
#DATA FORMATTING
df[['Quantity']]=df[['Quantity']].astype("int")
df['Date']=df['Date']+df['Time']
df['Date'] =  pd.to_datetime(df['Date'], format='%m/%d/%Y%H:%M')
df['Day'] = (df['Date']).dt.day
df['Month'] = (df['Date']).dt.month
df['Year'] = (df['Date']).dt.year
df['Hours'] = (df['Date']).dt.hour
df['Minutes'] = (df['Date']).dt.minute


#CHANGING THE TYPES OF COLUMNS 
df[['Day']]=df[['Day']].astype("int")
df[['Month']]=df[['Month']].astype("int")
df[['Year']]=df[['Year']].astype("int")
df[['Hours']]=df[['Hours']].astype("int")
df[['Minutes']]=df[['Minutes']].astype("int")

df=df.rename(columns={"Date":"DateTime"})
```

```python
df.dtypes
```

Now the data looks more approachable.

Let us also round up the floating values for easier accessibility and calculation.

```python
df=df.round({'Unit price': 2, 'Tax 5%': 2,'Total':2, 'cogs':2, 'gross margin percentage':2,
          'gross income':2,'rating':1})
df.head()
```

# Binning

For a clear usage of the price & rating, we will perform binning on both these features.

For Price, we are going to create three categories --- LOW, MEDIUM, HIGH

For Rating, we are going to create five categories --- VERY LOW, LOW, MEDIUM, HIGH, VERY HIGH

Let's perform this and create seperate columns & add them to our working dataframe

```python
#BINNING
bins = np.linspace(min(df["Total"]), max(df["Total"]), 4)


bin_names=['Low','Medium','High']
df['Price Category'] = pd.cut(df['Total'], bins, labels=bin_names, include_lowest=True)
#print(df['Price Category','Total'].head())


bins1 = np.linspace(min(df["Rating"]), max(df["Rating"]), 6)

bin_names1=['Very Low Rating','Low Rating','Medium Rating','High Rating','Very High Rating']
df['Rating Category'] = pd.cut(df['Rating'], bins1, labels=bin_names1, include_lowest=True)
#print(df['Rating Category','Rating'].head())


print(list(bins))

print(list(bins1))

df.head()
```

As observed in the list, The first list is the breakdown of binning of the price category for labelling.

The second list is the breakdown for the rating category.

Let us look at some insights of our data

```python
%matplotlib inline
def show_price_category():
    plt.pyplot.hist(df["Price Category"],color="#0504aa",alpha=0.50,bins='auto',rwidth=10)
    plt.pyplot.xlabel("Price Category")
    plt.pyplot.ylabel("Count")
    plt.pyplot.title("PRICE CATEGORY VS COUNT")
    plt.pyplot.show()

show_price_category()

```
